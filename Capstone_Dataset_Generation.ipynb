```python
#setting up a function to do jaccard similarity indexes by token between strings
def Jaccard(str1, str2):
    str1 = set(str1.split())
    str2 = set(str2.split())
    return float(len(str1 & str2)) / len(str1 | str2)
```

```python
#to keep track of the isbn when i find a match
isbns=[]
startpoint=0
endpoint=20
#this just makes it easy to run it on a partial dataset
test_case=df2#.iloc[startpoint:endpoint,:]

#making a list full of zeros of the appropriate length
for i in range(test_case.shape[0]):
    isbns.append(0)


#just grabbing the formatted title, year and supplier, also the actual title
for row in test_case.itertuples():
    #dont know if i'll need index at all
    index=row[0]
    title=row[1]
    year=row[2]
    supplier=row[3]
    #adding this so i can match even if the year is off
    real_title=row[4]
            
    #make the call with the formatted title
    resp = req.get("https://api2.isbndb.com/books/"+title+'?page=1&pageSize=40', headers=h)
    #default isbn to not found, NOT IMPLEMENTED YET
    isbn='Not Found'
    #keep track of jaccard similarity scores for publishers
    jaccards=[]
    #keep track of the date
    date=[]
    found_publisher=[]
    title_jaccards=[]
    found_titles=[]
    
    if resp.status_code==200:

        #need to add this in case there are no books returned for a search
        #just checks to make sure there is a 'books' field in the json, proceeds if there is
        book_check=[]
        for thing in resp.json():
            book_check.append(thing)
        if 'books' in book_check:
            #initiate a list of length equal to the number of books returned, all zeros, will index to appropriate point
            for i in range(len(resp.json()['books'])):
                date.append(0)
                found_titles.append(0)
                title_jaccards.append(0)
                found_publisher.append(0)
                jaccards.append(0)

            indexes_to_use=[]
            # iterate through the possible matches
            for idx, book in enumerate(resp.json()['books']):

                #create a list of keys from the dictionary

                keys=[]
                for key in book:
                    keys.append(key)

                if 'isbn13' in keys:
                    indexes_to_use.append(idx)
                    #checking title match
                title_jaccards[idx]=Jaccard(" ".join("".join([" " if ch in string.punctuation else ch for ch in book['title']]).split()).upper(),real_title)
                found_titles[idx]=book['title']
                #make sure the key exists before i call it
                #check the jaccard similarity score between the publisher listed and the supplier i have from diamond
                if 'publisher' in keys:
                    found_publisher[idx]=book['publisher']
                    jaccards[idx]=Jaccard(book['publisher'].upper(),supplier)
                #append -1 if there is no publisher
                #put the year in the appropriate index of the date list. 
                if 'date_published' in keys:
                    #need to use regex to extract YYYY dates. Format is messy
                    temp_date=re.findall('\d\d\d\d',book['date_published'])
                    if isinstance(temp_date,list):
                        #to deal with empty list
                        temp_date.append('error')
                        date[idx]=temp_date[0]
                    else:
                        date[idx]=temp_date
                elif 'publish_date' in keys:
                    temp_date=re.findall('\d\d\d\d',book['publish_date'])
                    if isinstance(temp_date,list):
                        temp_date.append('error')
                        date[idx]=temp_date[0]
                    else:
                        date[idx]=temp_date
                else:
                    date[idx]=0
                #NEED TO FIGURE OUT HOW TO FIND CORRECT MATCH AND APPEND THE ISBN13 to isbns LIST
            priority=9
            if len(indexes_to_use)>0:
                for i in indexes_to_use:
                    if jaccards[i] >=.5 and date[i]==year and title_jaccards[i]>=.5:
                        isbns[index]=resp.json()['books'][i]['isbn13']
                        print('good match')
                        priority=0
                        break
                    elif jaccards[i] >=.5 and title_jaccards[i]>=.65 and priority>1:
                        isbns[index]=resp.json()['books'][i]['isbn13']
                        print('publisher and title match')
                        priority=1
                    elif date[i] ==year and title_jaccards[i] >=.65 and priority>2:
                        isbns[index]=resp.json()['books'][i]['isbn13']
                        print('year and title match')
                        priority=2
                    elif jaccards[i] ==-1 and date[i]==year and priority>4:
                        isbns[index]=resp.json()['books'][i]['isbn13']
                        print('no supplier but year match')
                        priority=4
                    elif title_jaccards[i]>=.75 and priority >3:
                        isbns[index]=resp.json()['books'][i]['isbn13']
                        print('just title match')
                        priority=3


        #print(real_title,year,supplier,found_publisher,jaccards,date,found_titles,title_jaccards)
        print(index)
```
